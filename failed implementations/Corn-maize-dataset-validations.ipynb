```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

DATASET = "/root/.cache/kagglehub/datasets/smaranjitghose/corn-or-maize-leaf-disease-dataset/versions/1/data"

images = []

for cls in os.listdir(DATASET):
    cls_path = os.path.join(DATASET, cls)

    if os.path.isdir(cls_path):
        for img in os.listdir(cls_path):
            if img.lower().endswith((".jpg",".jpeg",".png")):
                images.append(os.path.join(cls_path,img))

sample = random.sample(images, 9)

plt.figure(figsize=(12,12))

for i,img_path in enumerate(sample):
    img = mpimg.imread(img_path)

    plt.subplot(3,3,i+1)
    plt.imshow(img)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
