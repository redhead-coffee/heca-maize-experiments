```python
#Load Classes
import os
import yaml

yaml_path = "/kaggle/input/datasets/yusufmurtaza01/tomato-leaf-disease/tomato/data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

class_names = data["names"]

NUM_CLASSES = len(class_names)

print("Classes:", NUM_CLASSES)

for i, c in enumerate(class_names):
    print(i, "->", c)
```

```python
#classification Dataset
import os

ROOT = "/kaggle/working/tomato_classification"

os.makedirs(ROOT, exist_ok=True)
os.makedirs(ROOT + "/train", exist_ok=True)
os.makedirs(ROOT + "/val", exist_ok=True)

print(ROOT)
```

```python
#convert YOLO Dataset to Classification Dataset
import shutil
from tqdm import tqdm

IMG_TRAIN = "/kaggle/input/datasets/yusufmurtaza01/tomato-leaf-disease/tomato/images/train"
IMG_VAL   = "/kaggle/input/datasets/yusufmurtaza01/tomato-leaf-disease/tomato/images/val"

LBL_TRAIN = "/kaggle/input/datasets/yusufmurtaza01/tomato-leaf-disease/tomato/labels/train"
LBL_VAL   = "/kaggle/input/datasets/yusufmurtaza01/tomato-leaf-disease/tomato/labels/val"

for cls in class_names:
    os.makedirs(os.path.join(ROOT, "train", cls), exist_ok=True)
    os.makedirs(os.path.join(ROOT, "val", cls), exist_ok=True)

# TRAIN
count_train = 0

for label_file in tqdm(os.listdir(LBL_TRAIN)):

    if not label_file.endswith(".txt"):
        continue

    with open(os.path.join(LBL_TRAIN, label_file), "r") as f:
        cls_id = int(f.readline().split()[0])

    cls_name = class_names[cls_id]

    img_name = label_file.replace(".txt", ".jpg")

    src = os.path.join(IMG_TRAIN, img_name)
    dst = os.path.join(ROOT, "train", cls_name, img_name)

    if os.path.exists(src):
        shutil.copy(src, dst)
        count_train += 1

print("train copied:", count_train)

# VAL
count_val = 0

for label_file in tqdm(os.listdir(LBL_VAL)):

    if not label_file.endswith(".txt"):
        continue

    with open(os.path.join(LBL_VAL, label_file), "r") as f:
        cls_id = int(f.readline().split()[0])

    cls_name = class_names[cls_id]

    img_name = label_file.replace(".txt", ".jpg")

    src = os.path.join(IMG_VAL, img_name)
    dst = os.path.join(ROOT, "val", cls_name, img_name)

    if os.path.exists(src):
        shutil.copy(src, dst)
        count_val += 1

print("val copied:", count_val)
print("CONVERSION COMPLETE")
```

```python
#validation Dataset
import tensorflow as tf

IMG_SIZE = (224,224)
BATCH_SIZE = 32

val_ds = tf.keras.utils.image_dataset_from_directory(
    "/kaggle/working/tomato_classification/val",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False,
    label_mode="int"
)

AUTOTUNE = tf.data.AUTOTUNE
val_ds = val_ds.prefetch(AUTOTUNE)

print("Validation Dataset Ready")
```

```python
class_names = sorted(os.listdir("/kaggle/working/tomato_classification/val"))

print(len(class_names))
print(class_names)
```

```python
#Load  Models
import tensorflow as tf

CA_MODEL_PATH = "/kaggle/input/models/redheadcoffee/heca-attention-models/keras/ca-heca-results/3/best_ca_model.keras"

HECA_MODEL_PATH = "/kaggle/input/models/redheadcoffee/heca-attention-models/keras/ca-heca-results/3/best_heca_model.keras"

print("Paths Found")
```

```python
import os

print("CA Exists :", os.path.exists(CA_MODEL_PATH))
print("HECA Exists :", os.path.exists(HECA_MODEL_PATH))
```

```python
#Load CA Model
ca_model = tf.keras.models.load_model(
    CA_MODEL_PATH,
    compile=False
)

print("CA Model Loaded")
```

Get Ground Truth Labels

```python
import numpy as np

y_true = np.concatenate([
    y.numpy()
    for x, y in val_ds
])

print("Ground Truth Shape:", y_true.shape)Get CA Probabilities
```

Get CA Probabilities

```python
ca_probs = ca_model.predict(
    val_ds,
    verbose=1
)

print(ca_probs.shape)
```

Generate Multiclass ROC Curve (CA)

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

y_true_bin = label_binarize(
    y_true,
    classes=np.arange(10)
)

plt.figure(figsize=(10,8))

for i in range(10):

    fpr, tpr, _ = roc_curve(
        y_true_bin[:, i],
        ca_probs[:, i]
    )

    roc_auc = auc(fpr, tpr)

    plt.plot(
        fpr,
        tpr,
        lw=2,
        label=f"{class_names[i]} (AUC={roc_auc:.3f})"
    )

plt.plot(
    [0,1],
    [0,1],
    linestyle="--"
)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - MobileNetV3 + Channel Attention")

plt.legend(
    fontsize=7,
    loc="lower right"
)

plt.grid(True)

plt.savefig(
    "/kaggle/working/ca_roc_curve.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import tensorflow as tf
import keras

keras.config.enable_unsafe_deserialization()

def spatial_avg(x):
    return tf.reduce_mean(x, axis=-1, keepdims=True)

def spatial_max(x):
    return tf.reduce_max(x, axis=-1, keepdims=True)

heca_model = tf.keras.models.load_model(
    HECA_MODEL_PATH,
    compile=False,
    safe_mode=False,
    custom_objects={
        "tf": tf,
        "spatial_avg": spatial_avg,
        "spatial_max": spatial_max
    }
)

print("HECA Model Reloaded")
```

```python
keras.config.enable_unsafe_deserialization()

heca_model = tf.keras.models.load_model(
    HECA_MODEL_PATH,
    compile=False,
    safe_mode=False
)

# Patch Lambda global namespace
for layer in heca_model.layers:
    if isinstance(layer, tf.keras.layers.Lambda):
        try:
            layer.function.__globals__["tf"] = tf
        except:
            pass

print("Patched")
```
